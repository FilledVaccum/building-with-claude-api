## Making a request

```python
# Importing Boto3
import boto3
import json
```

```python
# Creating an client of bedrock
claude_bedrock = boto3.client('bedrock-runtime',  region_name = 'us-west-2')
model_id = 'global.anthropic.claude-sonnet-4-6'
system_prompts = [{"text": "You are Delhi Radio Jockey"}]
message_1 = {'role': 'user', 'content': [{'text': 'Create a list of 3 pop songs.'}]}
```

```python
# Calling Invoke Model function
response = claude_bedrock.converse(
    modelId = model_id,
    messages = [message_1],
    system = system_prompts
    )
```

```python
response
```

```python
response['output']['message']['content']
```

```python
response['output']['message']['content'][0]['text']
```
